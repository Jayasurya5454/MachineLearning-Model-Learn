```python
import pandas as pd
```

```python
df=pd.read_csv("/content/cancer data.csv")
```

```python
df
```

```python
df.isnull().any()
```

```python
df.drop(columns=["id"],inplace = True)
```

```python
df
```

```python
df.drop(columns=["Unnamed: 32"],inplace = True)
```

```python
df
```

```python
x=df.iloc[:,1:]
```

```python
x
```

```python
y=df['diagnosis']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x,y,test_size = 25,random_state = 46)
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators = 6, criterion = 'entropy', random_state = 0)
classifier.fit(X_train, y_train)

y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
acc_logreg2 = round(accuracy_score(y_pred, y_test) , 2)*100
print("Accuracy : ",acc_logreg2)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,r2_score,classification_report
logreg = LogisticRegression(solver='lbfgs',max_iter=10000)
logreg.fit(X_train, y_train)
y_pred = logreg.predict(X_test)
acc_logreg1 = round(accuracy_score(y_pred, y_test) , 2)*100
print("Accuracy : ",acc_logreg1)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=5)
model=knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
acc_knn = round(accuracy_score(y_pred,y_test), 2) * 100
print("Accuracy :" ,acc_knn)
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report

# Create a Naive Bayes classifier
nb_classifier = MultinomialNB()

# Train the classifier with your training data
nb_classifier.fit(X_train, y_train)

# Make predictions on the test data
y_pred = nb_classifier.predict(X_test)

# Calculate accuracy and print the result
acc_nb = round(accuracy_score(y_pred, y_test) * 100, 2)
print("Accuracy : ", acc_nb)
```

```python
from sklearn.linear_model import SGDClassifier
from sklearn.metrics import accuracy_score, classification_report

# Create an SGDClassifier with appropriate settings
sgd_classifier = SGDClassifier(loss='log', max_iter=1000, random_state=42)

# Train the classifier with your training data
sgd_classifier.fit(X_train, y_train)

# Make predictions on the test data
y_pred = sgd_classifier.predict(X_test)

# Calculate accuracy and print the result
acc_sgd = round(accuracy_score(y_pred, y_test) * 100, 2)
print("Accuracy : ", acc_sgd)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report

# Create a Decision Tree classifier with appropriate settings
dt_classifier = DecisionTreeClassifier(random_state=42)

# Train the classifier with your training data
dt_classifier.fit(X_train, y_train)

# Make predictions on the test data
y_pred = dt_classifier.predict(X_test)

# Calculate accuracy and print the result
acc_dt = round(accuracy_score(y_pred, y_test) * 100, 2)
print("Accuracy: ", acc_dt)
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report

# Create an SVM classifier with appropriate settings
svm_classifier = SVC(kernel='linear', random_state=42)

# Train the classifier with your training data
svm_classifier.fit(X_train, y_train)

# Make predictions on the test data
y_pred = svm_classifier.predict(X_test)

# Calculate accuracy and print the result
acc_svm = round(accuracy_score(y_pred, y_test) * 100, 2)
print("Accuracy: ", acc_svm)
```
