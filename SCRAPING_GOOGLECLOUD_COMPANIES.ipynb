```python
import requests
from bs4 import BeautifulSoup
from googlesearch import search
import re
import pandas as pd

def get_company_website(company_name):
    try:
        query = f"official website of {company_name}"
        for result in search(query, num=1):
            return result
    except Exception as e:
        print(f"An error occurred: {e}")
        return None

url = "https://www.contino.io/insights/whos-using-google-cloud-platform"
response = requests.get(url)

if response.status_code == 200:
    soup = BeautifulSoup(response.text, 'html.parser')
    company_paragraphs = soup.find_all('p')
    start_printing = False
    company_data = []
    for paragraph in company_paragraphs:
        text = paragraph.get_text(strip=True)
        if "17 Media" in text:
            start_printing = True

        if start_printing:
            company_names = re.findall(r'\b[A-Za-z0-9 &\'"()/.-]+\b', text)

            for j, company_name in enumerate(company_names):
                if j != 200:
                    company_url = get_company_website(company_name)

                    if company_url is not None:
                        company_data.append({"company": company_name, "url": company_url})
                        print(f"{company_name}: {company_url}")

                    if j == 200:
                        break

    df = pd.DataFrame(company_data)
    df.to_csv("gcp_company.csv", index=False)
else:
    print(f"Failed to retrieve the webpage. Status code: {response.status_code}")
```
